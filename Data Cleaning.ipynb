### Remove blank rows and columns

```python
df_csv.isnull().all().sum()
```

```python
print(f"There are {df_csv.isnull().all(axis=1).sum()} blank rows in the dataset")
print(f"There are {df_csv.isnull().all(axis=0).sum()} blank columns in the dataset")
```
