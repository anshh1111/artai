```python
# Import the required libraries

import pandas as pd
import json
```

## Importing data of various formats

### CSV - Comma Separated Values

```python
df_csv = pd.read_csv('Global_Superstore2.csv', header=0, encoding='latin-1')
df_csv
```

### Xlsx - Excel Format

```python
df_excel = pd.read_excel('Global_Superstore2.xlsx')
df_excel
```

### XML - Extended Markup Language

```python
df_xml = pd.read_xml('books.xml')
df_xml
```

### JSON - Java Object Script Notation

```python
with open('sample.json') as f:
    contents = f.read()
data = json.loads(contents)
df_json = pd.json_normalize(data)
df_json
```
