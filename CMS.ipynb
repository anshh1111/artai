```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

# Load data from Excel file
excel_file = 'Career&Education.xlsx'  # Replace 'data.xlsx' with the path to your Excel file
df = pd.read_excel(excel_file)

# Create a new instance of the Chrome driver
driver = webdriver.Edge()

try:
    # Open the login page
    login_url = 'https://cms.compareiton.com/login'
    driver.get(login_url)

    # Login
    username = 'admin01@gmail.com'
    password = '123456789'
    username_input = driver.find_element(By.NAME, 'email')
    password_input = driver.find_element(By.NAME, 'password')
    username_input.send_keys(username)
    password_input.send_keys(password)
    password_input.send_keys(Keys.RETURN)

    # Wait for login
  # Adjust the wait time as needed

    # Iterate over each row in the Excel file
    for index, row in df.iterrows():
        # Navigate to the "Create Posts" page
        create_posts_url = 'https://cms.compareiton.com/posts'
        driver.get(create_posts_url) 
        try:
            # Find and fill in the form fields
            # Find the title input element using XPath
            title_input = driver.find_element(By.XPATH, "//label[text()='Title']/following-sibling::input")
            title_input.clear()
            title_input.send_keys(row['Title'])


            
            # Define the category to select
# Define the category to select
            selected_category = 'Travel'

# Select the category dropdown
            category_select = driver.find_element(By.NAME, 'category')
            category_select.click()

# Find all option elements within the dropdown
            category_options = category_select.find_elements(By.TAG_NAME, 'option')

# Iterate over each option to find the desired category
            for option in category_options:
            
                if option.text == selected_category:
                
        # Scroll to the option to ensure it's visible
                    driver.execute_script("arguments[0].scrollIntoView(true);", option)
        
        # Click the option
                    option.click()
                    break
            else:
                print(f"Error: Category '{selected_category}' not found in the dropdown.")



            # Define the selected value for trending (e.g., 'No' or '0')
            selected_trending = 'No'

# Select the trending dropdown
            trending_select = driver.find_element(By.NAME, 'trending')
            trending_select.click()

# Find all option elements within the dropdown
            trending_options = trending_select.find_elements(By.TAG_NAME, 'option')

# Flag to check if the trending option is found
            trending_found = False

# Iterate over each option to find the desired trending option
            for option in trending_options:
                if option.text.lower() == selected_trending.lower():
        # Scroll to the option to ensure it's visible
                    driver.execute_script("arguments[0].scrollIntoView(true);", option)
        
        # Click the option
                    option.click()
                    trending_found = True
                    break

# If the trending option is not found, print an error message
            if not trending_found:
                print(f"Error: Trending option '{selected_trending}' not found in the dropdown.")
                   
            selected_feature = 'No'

# Select the trending dropdown
            feature_select = driver.find_element(By.NAME, 'feature')
            feature_select.click()

# Find all option elements within the dropdown
            feature_options = feature_select.find_elements(By.TAG_NAME, 'option')

# Flag to check if the trending option is found
            feature_found = False

# Iterate over each option to find the desired trending option
            for option in feature_options:
                if option.text.lower() == selected_feature.lower():
        # Scroll to the option to ensure it's visible
                    driver.execute_script("arguments[0].scrollIntoView(true);", option)
        
        # Click the option
                    option.click()
                    feature_found = True
                    break

# If the trending option is not found, print an error message
            elif not feature_found:
                print(f"Error: Trending option '{selected_feature}' not found in the dropdown.")
 
            # Repeat the process for other form fields...
           # Find the input field for Meta based on the adjacent label
           meta_input = driver.find_element(By.XPATH, "//label[text()='Meta']/following-sibling::input")

# Clear the existing value in the Meta input field
           meta_input.clear()

# Enter the Meta value from the DataFrame row
           meta_input.send_keys(row['Meta'])

    
    
            # Submit the form
            submit_button = driver.find_element(By.CSS_SELECTOR, 'button[type="submit"]')
            submit_button.click()

            print(f"Post {index + 1} created successfully.")

        except Exception as e:
            print(f"Error creating post {index + 1}: {e}")

except Exception as e:
    print("An error occurred during login:", e)

finally:
    # Close the browser
    driver.quit()
```
