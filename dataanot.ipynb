```python
import numpy as np
from statsmodels.stats.porpotions import proportions_ztest

# Group A: Old website version
visitors_a = 100
conversions_a = 10

# Group B: New website version
visitors_b = 100
conversions_b = 18

# Combine data
successes = np.array([conversions_a, conversions_b])
trials = np.array([visitors_a, visitors_b])

# Run hypothesis test (Z-test for proportions)
stat, p_value = proportions_ztest(successes, trials)

# Print results
print("Z-statistic:", round(stat, 2))
print("p-value:", round(p_value, 4))

# Interpret result
alpha = 0.05
if p_value < alpha:
    print("✅ Reject Null Hypothesis: New design improved sales.")
else:
    print("❌ Fail to Reject Null: Not enough evidence to say new design helped.")
```
