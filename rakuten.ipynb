```python
import requests
import os

# Replace with the URL of the webpage containing the images
url = "https://adstransparency.google.com/advertiser/AR00138134394576568321?region=anywhere"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract the HTML content from the response
    html_content = response.text
    
    # Use a library like BeautifulSoup to parse the HTML content
    # and extract the image URLs
    # Example using BeautifulSoup:
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(html_content, "html.parser")
    
    # Find all image tags and extract the 'src' attribute
    image_tags = soup.find_all("img")
    
    # Create a directory to store the downloaded images
    if not os.path.exists("images"):
        os.makedirs("images")
    
    # Download each image
    for img in image_tags:
        # Get the URL of the image
        image_url = img.get("src")
        
        # Send a GET request to download the image
        image_response = requests.get(image_url)
        
        # Generate a unique filename for each image
        image_filename = os.path.join("images", os.path.basename(image_url))
        
        # Save the image to disk
        with open(image_filename, "wb") as f:
            f.write(image_response.content)
        
        print(f"Downloaded image: {image_filename}")
else:
    print(f"Request failed with status code {response.status_code}.")
```

```python
import requests
from bs4 import BeautifulSoup
import os

# Replace with the URL of the webpage containing the images
url = "https://adstransparency.google.com/advertiser/AR00138134394576568321?region=anywhere"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract the HTML content from the response
    html_content = response.text
    
    # Use BeautifulSoup to parse the HTML content
    soup = BeautifulSoup(html_content, "html.parser")
    
    # Find all image tags and extract the 'src' attribute
    image_tags = soup.find_all("img")
    
    # Create a directory to store the downloaded images
    if not os.path.exists("images"):
        os.makedirs("images")
    
    # Download each image
    for img in image_tags:
        # Get the URL of the image
        image_url = img.get("src")
        
        # Skip if the image URL is empty
        if not image_url:
            continue
        
        # Send a GET request to download the image
        image_response = requests.get(image_url)
        
        # Generate a unique filename for each image
        image_filename = os.path.join("images", os.path.basename(image_url))
        
        # Save the image to disk
        with open(image_filename, "wb") as f:
            f.write(image_response.content)
        
        print(f"Downloaded image: {image_filename}")
else:
    print(f"Request failed with status code {response.status_code}.")
```

```python
import requests
from bs4 import BeautifulSoup
import os
import urllib

# Replace with the URL of the website you want to scrape
url = "https://tpc.googlesyndication.com/archive/simgad/17384135024671910523"

# Send a GET request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Extract the HTML content from the response
    html_content = response.text
    
    # Use BeautifulSoup to parse the HTML content
    soup = BeautifulSoup(html_content, "html.parser")
    
    # Find all image tags and extract the 'src' attribute
    image_tags = soup.find_all("img")
    
    # Create a directory to store the downloaded images
    if not os.path.exists("images"):
        os.makedirs("images")
    
    # Download each image
    for img in image_tags:
        # Get the URL of the image
        image_url = img.get("src")
        
        # Skip if the image URL is empty
        if not image_url:
            continue
        
        # Generate a unique filename for each image
        image_filename = os.path.join("images", os.path.basename(image_url))
        
        # Download the image
        urllib.request.urlretrieve(urllib.parse.urljoin(url, image_url), image_filename)
        
        print(f"Downloaded image: {image_filename}")
else:
    print(f"Request failed with status code {response.status_code}.")
```
