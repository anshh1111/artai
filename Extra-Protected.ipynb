```python
clone_bot.rename(columns={'blocked_ip': 'user_ip'}, inplace=True)
```

```python

# Convert camp_id columns to the same type if needed (e.g., string)
clone_bot['camp_id'] = clone_bot['camp_id'].astype(str)
clone_camp['camp_id'] = clone_camp['camp_id'].astype(str)

# Merge the two dataframes on 'camp_id' and 'user_ip'
merged_df = pd.merge(clone_bot, clone_camp, on=['camp_id', 'user_ip'], how='inner')

# Extract the 'user_ip' column from the merged dataframe
common_user_ips = merged_df['user_ip'].unique()

# Display the list of user IPs common in both dataframes
print(common_user_ips)

```
